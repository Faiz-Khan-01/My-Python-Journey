Match Case Statements

```python
number = int(input("enter a number between 0-5:\t"))

# now it will give the output of the matched case 
match number :
    case 0 :
        print('You got a Nissan')
    case 1 :
        print('You got a BMW')
    case 2 :
        print('You got a Toyota')
    case 3 :
        print('You got an Audi')
    case 4 :
        print('You got a Mercedes')
    case 5 :
        print('You got a Honda')
    case _ if (int(number) > 5) :
        print('You got a Tesla')
```
