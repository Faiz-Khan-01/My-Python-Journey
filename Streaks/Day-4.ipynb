```python
# String Slicing

name = "Faiz"
print(name[1:3])
print(name[-3:-1]) # negative slicing
```

```python
# Advanced String Slicing

Fruit = "Pineapple"

print(Fruit[1:4:2]) # Start:Stop:Step 
print(Fruit[1:]) # the second place will take the last value
print(Fruit[:6]) # the first place blank will take first value
print(Fruit[::3])
```

```python
# Strings Functions

Country = "germany"

print(Country.endswith('any')) # checks if it ends with this
print(Country.count('P')) # counts number of appearance of given value
print(Country.capitalize()) # capitalize first letter
print(Country.replace('r', 't ')) # replace the value with another
print(Country.upper()) # make all letter capital
print(Country.lower()) #make all letters small
print(Country.find('y')) # gives the number of the first occurance time of value
print(Country.index('a')) # same as find but gives error if not found
print(Country.isalnum()) # returns true if no character except alphabets and numbers is found
print(Country.isalpha()) # same as isalnum but only accepts alphabets for True 
print(Country.islower()) # true if all characters are lower
print(Country.isupper()) # true if all characters are capitalize
```

```python
# String Functions

sports = ' Martial Arts '
print(sports.strip()) # removes useless space from after and before string
print(sports.isprintable()) # returns false if it contains any letter that isn't printable like \n, \t
print(sports.isspace()) # true if it has only whitespace
print(sports.istitle()) #true if first letter is capitalized
print(sports.swapcase()) #swaps letters

sentence = 'He is a good guy......'
print(sentence.rstrip('.')) # removes the symbols FROM end
print(sentence.split(" ")) # makes list and removes the value in bracket
print(sentence.center(100)) # returns the string in center depending on the gap
print(sentence.startswith('She')) #checks if it starts with
print(sentence.title()) #converts into title line
```
