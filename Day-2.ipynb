```python
# Arithmetic Operators

Num_1 = 30
Num_2 = 20

print("20 + 30 is", Num_1 + Num_2) # Addition
print("30 - 20 is", Num_1 - Num_2)  # Subtraction
print("30 * 20 is", Num_1 * Num_2) # Multiplication
print("30/20 is", Num_1 / Num_2) # Devision
print("In 30/20, Remainder is", Num_1 % Num_2) # Modulus
print("30/20 without making remainder 0 is", Num_1 // Num_2) # Floor Division
print("30 with power of 5 is", Num_1 ** 5) # Exponentiation
```
