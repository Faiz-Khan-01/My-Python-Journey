```python
# Typecasting

number = 3
name = 'apple'

print(type(number)) # the current number is int
print(type(float(number))) # now it's a float
print(type(str(number))) # now it's a string
```

```python
# String Operations

car = "BMW F32" # Normal String
Bike = 'Aprilia RS 450' # We can close string with single quote too
Lines = '''\nThe Toyota AE86 is an iconic car. 
It has a 4 cylinder NA engine.
It came in Initial D.
It's a JDM legend.\n''' # We can have multiple lines as strings by '''

print(car, Bike)
print("He said to me,\"Are you Mad?\"") # to use quote in string, we use \
print("The Car is below is my Fav\n",car) # for new line in string, we use \n
print(Lines)

print(len(Bike)) # prints length of string
print(Bike[1], car[0], Bike[2]) # Strings Slicing
```

```python
# for Loop

games = ['Theotown', 'Fr legends', 'SSR']

for list in games : # using loop to list the Items
    print(list)

print('\n') # just to have gap between lines

for loop in range(11) : # using loop to count to 10 
    print(loop)
```
